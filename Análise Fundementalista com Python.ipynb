# 1. Coleta e Importação de Dados:
Para realizar uma análise fundamentalista, precisamos de dados financeiros detalhados. Além do yfinance, também podemos usar APIs de serviços como Alpha Vantage, Quandl, entre outros.

```python
import yfinance as yf
import pandas as pd


# Baixando dados históricos
data = yf.download('AAPL', start='2020-01-01', end='2025-01-01')

# Obtendo dados financeiros específicos da empresa
stock = yf.Ticker('AAPL')
financials = stock.financials
balance_sheet = stock.balance_sheet
cashflow = stock.cashflow
```

# Cálculo de Indicadores Financeiros:
Vamos calcular alguns indicadores importantes para a análise fundamentalista:

P/E Ratio (Preço/Lucro):

```python
pe_ratio = data['Close'] / stock.info['trailingEps']
pe_ratio
```

# Dividend Yield:

```python
dividend_yield = stock.info['dividendYield']
dividend_yield
```

# ROE (Retorno sobre o Patrimônio):

```python
roe = stock.info['returnOnEquity']
roe
```

Visualização de Dados:
A visualização é uma parte crucial da análise financeira, pois nos ajuda a identificar tendências e padrões.

# Gráfico de Fechamento de Preços:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(data.index, data['Close'])
plt.title('Fechamento da AAPL')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.grid()
plt.show()
```

# Distribuição dos Dividendos:

```python
dividends = stock.dividends

plt.figure(figsize=(10, 5))
plt.plot(dividends.index, dividends)
plt.title('Distribuição de Dividendos da AAPL')
plt.xlabel('Data')
plt.ylabel('Dividendo')
plt.grid()
plt.show()
```

# Análise de Crescimento:
Você pode usar técnicas de regressão para prever o crescimento futuro com base nos dados históricos:

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Preparando os dados
X = np.array([i for i in range(len(data['Close']))]).reshape(-1, 1)
y = data['Close'].values

# Treinando o modelo
model = LinearRegression()
model.fit(X, y)

# Fazendo previsões
future_days = 365
X_future = np.array([i for i in range(len(data['Close']) + future_days)]).reshape(-1, 1)
future_pred = model.predict(X_future)

# Visualização
plt.figure(figsize=(10, 5))
plt.plot(data.index, data['Close'], label='Histórico de Fechamento')
plt.plot(pd.date_range(start=data.index[-1], periods=future_days, freq='D'), future_pred[-future_days:], label='Previsão de Fechamento')
plt.legend()
plt.title('Previsão de Crescimento dos Lucros da AAPL')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.grid()
plt.show()
```

```python

# Baixando dados históricos
data = yf.download('AAPL', start='2020-01-01', end='2025-01-01')

# Obtendo informações detalhadas da empresa
stock = yf.Ticker('AAPL')

# Calculando o P/E ratio
pe_ratio = data['Close'] / stock.info['trailingEps']

print(pe_ratio)
```
